```python
import pandas as pd

df = pd.read_csv('student-mat.csv', sep=';')
```

```python
df[['studytime', 'G3']].head()
```

```python
import matplotlib.pyplot as plt
plt.scatter(df['studytime'], df['G3'])
plt.xlabel('Study Time')
plt.ylabel('Final Grade (G3)')
plt.title('Study Time vs Final Grade')
plt.show()
```

This scatter plot shows the relationship between studentsâ€™ study time and their final grades (G3).

From the plot, it can be seen that students with higher study time tend to have slightly higher grades, but the pattern is not very strong.  
Students with the same study time still have a wide range of final grades.

This indicates that while study time may have some effect on performance, other factors also play an important role.

```python
avg_gender = df.groupby('sex')['G3'].mean()
avg_gender.plot(kind='bar')
plt.xlabel('Gender')
plt.ylabel('Average Final Grade (G3)')
plt.title('Average Final Grade by Gender')
plt.show()

```

The bar chart helps compare the final grades of male and female students.
Male students have a slightly higher average score.
Overall, the performance difference between genders is small.
